```python
merged_df23 = pd.merge(df2, df3, on='tconst')
merged_df23.head()
```

```python
merged_df23.info()
```

```python
merged_df23= merged_df23.drop(columns=['original_title'])
```

```python
merged_df23 = merged_df23.rename(columns={'primary_title': 'title'})
```

```python
merged_df23.head()
```

```python
merged_df23.isna().sum()
```

```python
merged_df23 = merged_df23.assign(genres=merged_df23['genres'].str.split(',')).explode('genres')
merged_df23.head(10)
```

```python
merged_df23.info()
```

```python
merged_df23.isna().sum()
```

```python
#rename df4 column data
df4 = df4.rename(columns={'movie': 'title'})
df4.head(100)
```

```python
df4.info()
```

```python
#merged_df23[merged_df23['title'] == 'Foodfight!']
```

```python
merged_df23.info()
```

```python
#merge df2 &df4
merged_df2df4 = pd.merge(merged_df23, df4, on='title')

merged_df2df4.head(100)
```

```python
merged_df2df4.info()
```

```python
merged_df2df4.isna().sum()
```

```python
df5=merged_df2df4.dropna()
df5.isna().sum()
```

```python
df5.head()
```

```python
df5['worldwide_gross'] = df5['worldwide_gross'].replace('[\$,]', '', regex=True).astype(float)
df5['domestic_gross'] = df5['domestic_gross'].replace('[\$,]', '', regex=True).astype(float)
df5['production_budget'] = df5['production_budget'].replace('[\$,]', '', regex=True).astype(float)
```

```python
df5.info()
```

```python
df5.isna().sum()
```

```python
df5.head()
```

```python
df5['production_budget'].max()
```

```python
#box office to million
df5['box office']= (df5['worldwide_gross']+ df5['domestic_gross'])
df5['box office']=round(df5['box office']/1e+06,1)
```

```python
#production buget to million
df5['production_budget']=round(df5['production_budget']/1e+06,1)
```

```python
#bar graph between profit and genre
df5_grouped=df5.groupby('genres')['box office'].mean()
sorted_data = df5_grouped.sort_values(ascending=False)
sorted_data.head(10).plot(kind='bar', x='genres', y='box office',color='Green', figsize=(10, 6))
plt.title('BOX OFFICE VS GENRE')
plt.xlabel('Genre')
plt.ylabel('Box office(Million USD)')
plt.show()  
```
